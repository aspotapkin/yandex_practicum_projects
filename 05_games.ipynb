## Введение

Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Данные представлены до 2016 года. Необходимо спланировать кампанию на 2017-й. 
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».


**Описание данных:**
- *Name* — название игры
- *Platform* — платформа
- *Year_of_Release* — год выпуска
- *Genre* — жанр игры
- *NA_sales* — продажи в Северной Америке (миллионы проданных копий)
- *EU_sales* — продажи в Европе (миллионы проданных копий)
- *JP_sales* — продажи в Японии (миллионы проданных копий)
- *Other_sales* — продажи в других странах (миллионы проданных копий)
- *Critic_Score* — оценка критиков (максимум 100)
- *User_Score* — оценка пользователей (максимум 10)
- *Rating* — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

```python
#измерим размер исходного датафрейма
len_games_before = len(games)
```

```python
#приведение столбцов к нижнему регистру
games.columns = games.columns.str.lower()
games.columns
```

Большое количество данных пропущено в графах с оценками критиков, пользователей и рейтинге ERSB. В двух случаях неизвестно название и жанр игры, в 269 строках отсутствует год выпуска игры.

```python
#удаляем строки с отсутствующей информацией о годе и жанре игры
games = games.dropna(subset=['year_of_release', 'genre'])
#заглушка рейтинга ESRB - меняем на None
games.loc[games['rating'].isna(), 'rating'] = games.loc[games['rating'].isna(), 'rating'].fillna('Not defined')
games.isna().sum()
```

```python
#приводим значения года к типу дата
games['year_of_release'] = pd.to_datetime(games['year_of_release'].astype('int'), format='%Y')
# значения оценок критиков целочисленные, приведем значения оценок критиков к типу Int64 (работает с пустыми значениями)
games['critic_score'] = games['critic_score'].astype('Int64')
```

```python
games.info()
```

```python
#разбираемся с пропусками в оценках пользователей и критиков, а также рейтинге. Возможно, дело в том, что раньше такие значения просто не присваивались
ratings_year = (games.pivot_table(index='year_of_release', values = ['name', 'critic_score', 'user_score', 'rating'], 
                            aggfunc={'name': 'count','critic_score': 'count', 'user_score': 'count', 'rating': 'count'}).\
           reindex(['critic_score', 'user_score', 'rating', 'name'], axis=1).\
           reset_index())
ratings_year.columns = ['year', 'critic_scores', 'user_scores', 'ratings', 'total']
ratings_year.plot(x='year', y=['critic_scores', 'user_scores', 'ratings', 'total'], kind='line')
```

Заметное количество игр с отзывами критиков и пользователей (100 и более) начало появляться только с 2000 года. При этом в дальнейшем общее число игр также превышало количество игр с оценками критиков и пользователей. Возможно, число оценок зависит от платформы. Также для игр, вышедших в 2016 г., рейтинги просто не успели определить.

```python
ratings_platform = (games.pivot_table(index='platform', values = ['name', 'critic_score', 'user_score', 'rating'], 
                            aggfunc={'name': 'count','critic_score': 'count', 'user_score': 'count', 'rating': 'count'}).\
           reindex(['critic_score', 'user_score', 'rating', 'name'], axis=1))
ratings_platform.columns = ['critic_scores', 'user_scores', 'ratings', 'total']
#добавляем доли игр с оценками от общего количества игр, выпущенны
ratings_platform['cs_share'] = round(ratings_platform['critic_scores'] / ratings_platform['total'] * 100, 1)
ratings_platform['us_share'] = round(ratings_platform['user_scores'] / ratings_platform['total'] * 100, 1)
ratings_platform['rating_share'] =  round(ratings_platform['ratings'] / ratings_platform['total'] * 100, 1)
ratings_platform.sort_values('us_share', ascending=False).reset_index()
```

Только 4 платформы, представленные в датасете, имеют долю игр с оценками критиков, пользователей и рейтинга ERSB больше 70%. Удаление и замена средними значениями не представляется возможным, поскольку на оценки может повлиять большое количество факторов.

```python
games['total_sales'] = games['na_sales'] + games['eu_sales'] + games['jp_sales'] + games['other_sales']
games = games.reindex(['name', 'platform', 'year_of_release', 'genre', 'na_sales', 'eu_sales',
       'jp_sales', 'other_sales', 'total_sales', 'critic_score', 'user_score', 'rating'], axis=1)
```

```python
#аббревиатура 'tbd'
tbd_games = games.loc[games['user_score'] == 'tbd']
tbd_games.info()
```

*tbd (to be determined)* - рейтинг не определен. Посмотрим на распределение данных.

```python
tbd_games.hist(figsize=(18, 20))
```

Большинство таких игр имеют небольшое количество продаж. Возможно, игры, выпущенные до 2016 г. не понравились пользователям. По графику с отзывами критиков видно, что такие игры получили более низкие оценки.

```python
#сравним средние оценки критиков
mean_critic_score_all = round(games['critic_score'].mean(), 2)
mean_critic_score_tbd = round(tbd_games['critic_score'].mean(), 2)
print(f'Средняя оценка критиков (все игры): {mean_critic_score_all}')
print(f'Средняя оценка критиков (пользовательский рейтинг не определен): {mean_critic_score_tbd}')
print(f'Разница: {round(mean_critic_score_all - mean_critic_score_tbd, 2)}')
```

```python
tbd_games['genre'].value_counts()
```

```python
tbd_games['platform'].value_counts()
```

```python
"""
Поскольку рейтинг у игр с небольшим числом продаж не определен, то заменять значения на средние и медианы не совсем корректно.
Поэтому предлагается заменить 'tbd' на NA
"""
#приводим значение пользовательского рейтинга в срезе данных к типу вещественного числа (float)
games[games['user_score'] == 'tbd'] = games[games['user_score'] == 'tbd'].replace('tbd', np.nan)
games['user_score'] = games['user_score'].astype('float')
    #итог
games.info()
```

Теперь можно пользовательский рейтинг приобрел числовой формат. 

```python
len_games_after = len(games)
processing_ratio = round(len_games_after / len_games_before, 2) 
print(f'Кол-во строк до обработки: {len_games_before}')
print(f'Кол-во строк после обработки: {len_games_after}')
print(f'% от исходного датафрейма: {processing_ratio * 100}')
```

**Итог:**
- удалены строки с пустыми значениями в названии игры и годе релиза
- добавлен столбец с общими продажами
- столбец с годом релиза приведен к типу datetime
- столбец с оценками критиков приведен к целочисленному значению
- столбец с пользовательскими оценками приведен к типу float, значения с неопределенным рейтингом заменены на средние значения пользовательского рейтинга игр со схожими продажами в зависимости от платформы
- от исходного датафрейма осталось 98%

Столбцы с региональными продажами имеют индекс 4, 5 и 6.

```python
def region_sales(data, param, top5=True):
    top_regions = []
    for reg in data.columns[4:7]:
        top_data = data.groupby(param)[reg].sum().sort_values(ascending=False)
        other = pd.Series(sum(top_data[5:]), index=['Other'])
        if len(top_data) > 5:
            top_5 = top_data.head().append(other)
            print(f'{reg}:\n', top_5)
            print('--------------')
            top_regions.append(top_5)
        else:
            print(f'{reg}:\n', top_data)
            print('--------------')
            top_regions.append(top_data)
        
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10,15), constrained_layout = True)
    ax1.pie(top_regions[0], labels=top_regions[0].index, autopct='%.0f%%')
    ax1.set_title(f'TOP 5 by {param} in North America')
    ax2.pie(top_regions[1], labels=top_regions[1].index, autopct='%.0f%%')
    ax2.set_title(f'TOP 5 by {param} in European Union')
    ax3.pie(top_regions[2], labels=top_regions[2].index, autopct='%.0f%%')
    ax3.set_title(f'TOP 5 by {param} in Japan')
```

### Проверка гипотез
- [x] Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- [x] Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Пороговое значение alpha = .05
Необходимо сформулировать нулевую и альтернативную гипотезы, а также обосновать выбрать критерия ее проверки.

**Гипотеза 1:**
- Н_0: Средние пользовательские рейтинги платформ Xbox One и PC равны
- Н_а: Средние пользовательские рейтинги платформ Xbox One и PC не равны

**Гипотеза 2:**
- Н_0: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») равны.
- Н_а: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») не равны.

Поскольку мы берем все игры за последние 5 лет, то нужно проверять гипотезы о равенстве средних двух генеральных совокупностей. Поскольку размер выборок разный, то параметр равенства выборок нужно указать *False*

```python
alpha = .05
sample_xbox = games_12_16.loc[games_12_16['platform'] == 'XOne', 'user_score'].dropna()
sample_pc = games_12_16.loc[games_12_16['platform'] == 'PC', 'user_score'].dropna()

h1 = st.ttest_ind(sample_xbox, sample_pc, equal_var = False)
print(f'p-value: {h1.pvalue}')
if h1.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
alpha = .05
sample_action = games_12_16.loc[games_12_16['genre'] == 'Action', 'user_score'].dropna()
sample_sports = games_12_16.loc[games_12_16['genre'] == 'Sports', 'user_score'].dropna()

h2 = st.ttest_ind(sample_action, sample_sports, equal_var = False)
print(f'p-value: {h2.pvalue}')
if h2.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```
